```python
"""
1. Histogram and fitting using Seaborn
=========================================
Plotting the histogram of sale price and fitting.
"""
# Authors: Muz Ahmad

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Comment this if the data visualisations doesn't work on your side
%matplotlib inline
plt.style.use('bmh')
df = pd.read_csv('house-price.csv')

# Remove Id and the features with 30% or less NaN values
df2 = df[[column for column in df if df[column].count() / len(df) >= 0.3]]
del df2['Id']
print("List of dropped columns:", end=" ")

for c in df.columns:
    if c not in df2.columns:
        print(c, end=", ")
print('\n')
df = df2
df.head()
```

```python
# Plot distribution of target feature (SalePrice)
print(df['SalePrice'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['SalePrice'], color='brown', bins=100, hist_kws={'alpha': 0.4});
```
