```python
# Try to fit data with different types of distribution by seaborn
y = df['SalePrice']
plt.figure(1); plt.title('Johnson SU distribution')
sns.distplot(y, kde=False, fit=st.johnsonsu, color='red')
plt.figure(2); plt.title('Normal distribution')
sns.distplot(y, kde=False, fit=st.norm, color='blue')
plt.figure(3); plt.title('Log Normal distribution')
sns.distplot(y, kde=False, fit=st.lognorm, color='green')
```
