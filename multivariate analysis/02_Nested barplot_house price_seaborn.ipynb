```python
"""
Nested barplot using Seaborn
=========================================
Plotting the nested barplot of sale price by House style and Land slope using Seaborn.
"""
# Authors: Muz Ahmad

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Comment this if the data visualisations doesn't work on your side
%matplotlib inline
plt.style.use('bmh')
df = pd.read_csv('house-price.csv')

# Remove Id and the features with 30% or less NaN values
df2 = df[[column for column in df if df[column].count() / len(df) >= 0.3]]
del df2['Id']
print("List of dropped columns:", end=" ")

for c in df.columns:
    if c not in df2.columns:
        print(c, end=", ")
print('\n')
df = df2
df.head()
```

```python
sns.set(style="whitegrid")

# Draw a nested barplot to show survival for class and sex
g = sns.catplot(x="HouseStyle", y="SalePrice", hue="LandSlope", data=df,
                height=6, kind="bar", palette="rainbow")
g.despine(left=True)
g.set_ylabels("Sale Price")
```
