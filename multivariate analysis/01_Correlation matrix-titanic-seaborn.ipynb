```python
"""
Correlation matrixusing Seaborn
==============================================================
Plotting the correlation matrix of all numerical features using Seaborn.
"""
# Authors: Muz Ahmad

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

# Read data from csv file
data=pd.read_csv('Titanic.csv')
data.head()
```

```python
# Drop features with non-numerical data
data.drop(['Name','Sex','Age','Ticket','Cabin','PassengerId'],axis=1,inplace=True)

# Plot correlation matrix
sns.heatmap(data.corr(),annot=True,cmap='RdYlGn',linewidths=0.2,annot_kws={'size':20})
fig=plt.gcf()
fig.set_size_inches(18,15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
